# Sonar Rock vs Mine Prediction

```python
#importing the libraries
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt

#Importing train test split
from sklearn.model_selection import train_test_split

#Importing classification algorithm
from sklearn.linear_model import LogisticRegression

#Importing evaluation metrices
from sklearn.metrics import accuracy_score
```

```python
import io
sonar_data = pd.read_csv(io.BytesIO(uploaded['Copy of sonar data.csv']), header=None)
```

```python
# separating data and Labels
X = sonar_data.drop(columns=60, axis=1)
Y = sonar_data[60]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.1, stratify=Y, random_state=42)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

## Logistic Regression

```python
model_LT = LogisticRegression()
```

```python
#training the Logistic Regression model with training data
model_LT.fit(X_train, Y_train)
```

Evaluation

```python
#accuracy on training data
X_train_prediction = model_LT.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train) 
```

```python
print('Accuracy on training data : ', training_data_accuracy)
```

```python
#accuracy on test data
X_test_prediction = model_LT.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test) 
```

```python
print('Accuracy on test data : ', test_data_accuracy)
```

```python
#Calculate the feature importance
importance = model_LT.coef_[0]

# Sort the features by their importance
sorted_indexes = importance.argsort()[::-1]

# Print the feature ranking
print("Feature ranking:")

for f in range(X.shape[1]):
    print("%d. feature %d (%f)" % (f + 1, sorted_indexes[f], importance[sorted_indexes[f]]))
```

```python
# Ploting the feature importance as a bar graph
plt.figure(figsize=(10, 5))
plt.bar(range(X.shape[1]), importance[sorted_indexes])
plt.xlabel("Feature")
plt.ylabel("Importance")
plt.title("Feature importance in Logistic Regression")
plt.show()
```
